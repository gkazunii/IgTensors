```julia
function binary_from_idx(idx, tensor_size)
    
     D = length(idx)
    bi = Vector{String}(undef,D)
    
    for d = 1:D
        logmode = Int( log2( tensor_size[d] ) )
        bi[d] = lpad(string(idx[d]-1,base=2), logmode, '0')
    end
    
    return bi
end
```
